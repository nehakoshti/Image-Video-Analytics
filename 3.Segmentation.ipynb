```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
image = cv2.imread('/content/drive/MyDrive/Trichy cllg sem 2 data/IVA Labs/3.butterfly.png')
```

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)     # convert the img from BGR to RGB (OpenCV loads images in BGR)
```

```python
# plt.imshow(image)

from google.colab.patches import cv2_imshow
st = np.hstack((image,image_rgb))
cv2_imshow(st)
```

```python





# Reshape the image to a 2D array of pixels (each pixel is a row with 3 color values)
pixels = image_rgb.reshape(-1, 3)

# Perform K-means clustering
k = 4  # number of clusters (segments)
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(pixels)

# Get the labels and centers
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# Map each pixel to its corresponding cluster center
segmented_image = centers[labels].reshape(image_rgb.shape).astype(int)

# Display the original and segmented images
plt.figure(figsize=(12, 6))

# Original image
plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis('off')

# Segmented image
plt.subplot(1, 2, 2)
plt.imshow(segmented_image)
plt.title(f"Segmented Image with {k} Clusters")
plt.axis('off')

plt.show()

# Optionally, save the segmented image
cv2.imwrite('segmented_image.jpg', cv2.cvtColor(segmented_image, cv2.COLOR_RGB2BGR))
```
